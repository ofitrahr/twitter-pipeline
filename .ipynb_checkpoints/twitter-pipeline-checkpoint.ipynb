```python
#STEP 1: IMPORT PACKAGES AND LIBRARIES

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import regex as re
import snscrape.modules.twitter as sntwitter
import stylecloud
from googletrans import Translator
from IPython.display import Image
from nltk.corpus import stopwords
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from textblob import TextBlob
from tqdm.notebook import tqdm
from wordcloud import WordCloud
from PIL import Image
```

```python
#STEP 2: SCRAPING THE DATA

query ='turkiye'
scraper = sntwitter.TwitterSearchScraper(query)
tweets = []
limit = 100

for i, tweet in enumerate(scraper.get_items()):
    
    data = [
        tweet.date,
        tweet.user.username,
        tweet.user.id,
        tweet.user.verified,
        tweet.user.created,
        tweet.user.followersCount,
        tweet.user.friendsCount,
        tweet.user.statusesCount,
        tweet.user.location,
        tweet.replyCount,
        tweet.retweetCount,
        tweet.likeCount,
        tweet.lang,
        tweet.coordinates,
        tweet.viewCount,
        tweet.renderedContent
        
    ]
    
    tweets.append(data)  
    
    if i > limit:
        break
        
tweets_df = pd.DataFrame(tweets, columns = ['Date',
                                            'Username',
                                            'User_id',
                                            'User_verified',
                                            'User_creation_date',
                                            'User_Followers',
                                            'User_Friends',
                                            'User_status_numbers',
                                            'User_location',
                                            'ReplyCount',
                                            'RetweetCount',
                                            'LikeCount',
                                            'Lang',
                                            'Coordinates',
                                            'ViewCount',
                                            'Content'
                                           ])

print(tweets_df)
```

```python
#STEP 3: CLEANING THE TWEETS

def clean_tweet(tweet):
    
    ##Lowering all the letters
    temp = tweet.lower()

    ##Removing hashtags and mentions
    temp = re.sub("@[A-Za-z0-9_]+","", temp)
    temp = re.sub("#[A-Za-z0-9_]+","", temp)

    ##Removing links
    temp = re.sub(r"http\S+", "", temp)
    temp = re.sub(r"www.\S+", "", temp)

    ##Removing punctuations
    temp = re.sub('[()!?]', ' ', temp)
    temp = re.sub('\[.*?\]',' ', temp)

    ##Filtering non-alphanumeric characters
    temp = re.sub("[^a-z0-9]"," ", temp)
    temp = re.sub(r'[0-9]+','', temp)              
        
    ##Translating multiple language into single lang
    translater = Translator()
    out = translater.translate(temp, dest='en')
    temp = out.text
    
    ##Tokenization
    temp = temp.split()
    
    ##Stopwords removal
    stopword = stopwords.words('english')
    temp = [w for w in temp if not w in stopword]
    temp = " ".join(word for word in temp)

    return temp

tweets_df['Cleaned_Content'] = [clean_tweet(tweet) for tweet in tweets_df['Content']]
tweets_df.drop('Content', inplace=True, axis=1)
tweets_df.to_csv('cleaned_tweets_python.csv')

print(tweets_df)
```

```python
#STEP 4: MAKE A WORDCLOUD FROM CLEANED TWEETS

tweets_df['Cleaned_Content'].to_csv('tweets_only.csv', index=False)

stylecloud.gen_stylecloud(file_path='tweets_only.csv',
                          icon_name='fas fa-leaf',
                          palette='cartocolors.diverging.ArmyRose_2',
                          background_color='black',
                          gradient='horizontal',
                          stopwords=True,
                          custom_stopwords= ['said','vndb','org v'])

Image('stylecloud.png')
```

```python
#STEP 5: BASIC SENTIMENT ANALYSIS

tweets_df['sentiment_score'] = [TextBlob(tweet).sentiment.polarity for tweet in tweets_df['Cleaned_Content']]

for i, tweet in enumerate(tweets_df['sentiment_score']):
    
    if tweet > 0:
        tweets_df.loc[i,'sentiment'] = 'Positive'
     
    elif tweet == 0:
        tweets_df.loc[i,'sentiment'] = 'Neutral'
        
    else:
        tweets_df.loc[i,'sentiment'] = 'Negative'

print(tweets_df[['Cleaned_Content','sentiment_score','sentiment']])
print(tweets_df['sentiment'].value_counts().plot(kind='bar'))
```

```python
#STEP 6: SENTIMENT ANALYSIS VISUALIZATION IN WORDCLOUD

#Change the text dataframe into one single list
positive_text = ' '.join(tweets_df[tweets_df['sentiment_score'] > 0]['Cleaned_Content'])
negative_text = ' '.join(tweets_df[tweets_df['sentiment_score'] < 0]['Cleaned_Content'])
all_text = positive_text + ' ' + negative_text

positive_word_frequencies = {word: positive_text.count(word) for word in positive_text.split()}
negative_word_frequencies = {word: negative_text.count(word) for word in negative_text.split()}
all_word_frequencies = {word: all_text.count(word) for word in all_text.split()}

# Create a custom color function that assigns different colors to positive and negative words
def color_func(word, font_size, position, orientation, random_state=None, **kwargs):
    if word in positive_word_frequencies:
        return 'green'
    elif word in negative_word_frequencies:
        return 'red'
    else:
        return 'gray'

wordcloud = WordCloud(background_color='white', color_func=color_func, mask=np.array(Image.open('cloud.png'))).generate_from_frequencies(all_word_frequencies)

plt.figure(figsize=(10, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
#STEP 7: SENTIMENT ANALYSIS VISUALIZATION IN BAR CHART

counts = tweets_df.groupby('Lang')['sentiment'].value_counts().reset_index(name='count')

fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

counts.query("Lang == 'de'").plot(x='sentiment', y='count', kind='bar', ax=ax1, color='purple', alpha=0.7)
counts.query("Lang == 'en'").plot(x='sentiment', y='count', kind='bar', ax=ax2, color='orange', alpha=0.7)

ax1.set_xlabel('Unique Names')
ax1.set_ylabel('Count (female)', color='purple')
ax2.set_ylabel('Count (male)', color='orange')
plt.title('Frequency of Unique Names by Gender')

# show the plot
plt.show()
```
